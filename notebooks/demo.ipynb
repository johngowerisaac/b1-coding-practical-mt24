```python
import sys
import os

# Add the path to the uuv_mission directory without moving files
uuv_mission_path = os.path.abspath('../uuv_mission')  # This will go up one directory and then into uuv_mission
sys.path.append(uuv_mission_path)

# Now import specific classes from dynamic
from dynamic import Submarine, Mission, ClosedLoop
from control import PDController

# Print current working directory for debugging
# print("Current Working Directory:", os.getcwd())

# Verify the path to the CSV fileff
# csv_path = '../data/mission.csv'
# print("Does the CSV file exist?", os.path.exists(csv_path))

# Instantiate your controller (depending on your implementation)
submarine = Submarine()
mission = Mission.from_csv()
controller = PDController(Kp=0.15, Kd=0.6)

# Create a ClosedLoop instance
closed_loop = ClosedLoop(submarine, controller)

trajectory = closed_loop.simulate_with_random_disturbances(mission)
trajectory.plot_completed_mission(mission)
```
